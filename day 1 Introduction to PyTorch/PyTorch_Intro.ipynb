### Inplace operations

```python
m = torch.rand(2, 3)
n = torch.rand(2, 3)
# so if we try to add these 2 new memory space will get created for new output tensor
# to reduce this extra space we use inplace operations

m.add_(n)

# now value is stores in m itself so no extra space is occupies
```

### Let's test why GPU and not CPU

```python
import time

size = 10000

# time on CPU
m1_cpu = torch.rand(size, size)
m2_cpu = torch.rand(size, size)

start_time = time.time()
res = torch.matmul(m1_cpu, m2_cpu)
cpu_time = time.time() - start_time

print(f"CPU time: {cpu_time:.4f} seconds")

# time on GPU

m1_gpu = torch.rand(size, size, device='cuda')
m2_gpu = torch.rand(size, size, device='cuda')

start_time = time.time()
res = torch.matmul(m1_gpu, m2_gpu)
gpu_time = time.time() - start_time

print(f'GPU time: {gpu_time:.4f} seconds')

print(f'GPU is {cpu_time/gpu_time:.4f} times faster than CPU')
```
